```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancerdata = load_breast_cancer()
cancerdata.keys()
```

```python
print(cancerdata['DESCR'])
```

```python
cancerdata['data'].shape, type(cancerdata['data'])
```

```python
df = pd.DataFrame(np.c_[cancerdata['data'],cancerdata['target']], columns = np.append(cancerdata['feature_names'],['target_names']))
df.head()
```

```python
import seaborn as sns
sns.pairplot(df, hue='target_names', vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness','mean concavity','mean symmetry'])
```
