```python
df = pd.read_csv("appdata10.csv")
```

```python
df.head()
```

```python
y=df['enrolled']
df=df.drop(['enrolled'],axis=1)
from  sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df,y,test_size=0.4,random_state=69)
```

```python
type(X_train)
```

```python
train = pd.concat([X_train,y_train],axis=1)
train.reset_index(drop=True)
```

```python
test = pd.concat([X_test,y_test],axis=1)
test.reset_index(drop=True)
```

```python
train.shape,test.shape
```

```python
sample = test[['user','enrolled']].iloc[:5]
actualresult = test[['user','enrolled']]
```

```python
test['enrolled'] = np.nan
```
